```python
loss = []
acc = []
with open("./dataset/result_small.txt",'r') as file:
    temp_loss = 0
    temp_acc = 0
    num = 0
    for i in file.readlines():
        if "~~~~~~~~~~" in i :
            if num!=0 :                
                loss.append(temp_loss/num)
                acc.append(temp_acc/num)
                temp_loss = 0
                temp_acc = 0
                num = 0
        if "loss" in i:
            num+=1
            temp_loss += float(i.split("loss:")[1].split(" ")[1])
            temp_acc += float(i.split("loss:")[1].split(" ")[-1])
        print(i)
    print(loss)
    print(acc)
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1,len(loss)+1), loss, '-o')
plt.xlabel("epoch")
plt.ylabel("loss")
plt.show()

plt.plot(range(1,len(acc)+1), acc, '-o')
plt.xlabel("epoch")
plt.ylabel("acc")
plt.show()
```

```python
def triplet_loss(y_true, y_pred):
    margin = K.constant(4)
    return K.mean(K.maximum(K.constant(0), y_pred[:,0,0] - y_pred[:,1,0] + margin))
```

```python
def euclidean_distance(vects):
    x, y = vects
#     return K.sqrt(K.maximum(K.sum(K.square(x - y), axis=1, keepdims=True), K.epsilon()))
    return  K.sqrt(K.sum(K.square(x - y), axis=1, keepdims=True))
```

```python
def myAccuracy(y_true, y_pred):
    return K.mean(y_pred[:,0,0] < y_pred[:,1,0])
```
